```python
import tweepy
import pandas as pd
import time
consumer_key = ""
consumer_secret = ""
access_token = ""
access_token_secret = ""
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth,wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
```

```python
df=pd.read_csv(r"D:\webscraping\twitter_scraping.csv")
```

```python
following = []
description = []
name = []
location = []
followers_count = []
verified = []
protected = []
statuses_count = []
```

```python
i=0
backoff_counter=1
following=[]
def get_following(screen_name,i,backoff_counter):
    if i==100:
        return following
    else:
        while True:
            try:
                for friends in tweepy.Cursor(api.friends, screen_name).items(100):
                    if friends.screen_name not in following:
                        following.append(friends.screen_name)
                        description.append(friends.description)
                        name.append(friends.name)
                        followers_count.append(friends.followers_count)
                        verified.append(friends.verified)
                        protected.append(friends.protected)
                        location.append(friends.location)
                        statuses_count.append(friends.statuses_count)
                return get_following(following[i], i+1,backoff_counter)
            except tweepy.TweepError as e:
                print(e.reason)
                time.sleep(60*backoff_counter)
                backoff_counter += 1
                continue
```

```python
following, description, name, following, followers_count, verified, protected, location, statuses_count = get_following('viraj_sheth', i,backoff_counter=1)
print(len(following))
```

```python
data={'Name':name,'Followers_count':followers_count,'Verified':verified,'Protected':protected,'Location':location,'Status_count':statuses_count,'Description':description}
df=pd.DataFrame(data)
```
